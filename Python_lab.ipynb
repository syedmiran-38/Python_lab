**1) Write a python program which prompts user to enter the radius of a circle  if the radius “ i” greater than “0” the calculate & print the area , circumference of the circle.**

```python
radius = float(input("Enter the radius of the circle: "))

if radius > 0:
    area = 3.14 * radius ** 2
    circumference = 2 * 3.14 * radius
    print("The area of the circle is:", round(area, 2))
    print("The circumference of the circle is:", round(circumference, 2))
else:
    print("Invalid radius. Please enter a radius greater than 0.")
```

**1) Write a python program to calculate the factorial of a given number using while loop. read the number from keyword**

```python
num = int(input("Enter a number: "))
factorial = 1

while num > 0:
    factorial = factorial * num
    num = num - 1

print(f"The factorial of the given number is: {factorial}")
```

**2) Write a python program to find the sum of the digits of a given number**

```python
num = int(input("Enter a number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num = num // 10

print(f"The sum of the digits of the given number is: {sum}")
```

**3) What is lambda function ?what are the characteristics of a lambda function?**

A lambda function in Python is a small, anonymous, and inline function. It is defined using the lambda keyword and can have any number of arguments, but it can only have one expression. The expression is executed when the lambda function is called.
Some of the characteristics of a lambda function are:

Anonymous: A lambda function doesn't have a name and is defined using the lambda keyword.

Inline: A lambda function is defined in a single line of code and can be written in a compact form.

Flexible: A lambda function can take any number of arguments, but it can only have one expression.

Used as argument: A lambda function can be passed as an argument to another function.

**1) Write a python program to display the reverse of a given number**

```python
num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    last_digit = num % 10
    reverse = reverse * 10 + last_digit
    num = num // 10

print("The reverse of the given number is:", reverse)
```

**2) Write a python program to check whether the number entered is Armstrong number**

```python
num = int(input("Enter a number: "))
sum = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum += digit**3
    temp = temp // 10

if num == sum:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

**1) Write a python program to print Fibonacci series up to 8th term**

```python
a = 0
b = 1
n = 8

print('The fibonacci series upto 8th term is = ')
print(a)
print(b)

for i in range(2, n):
    c = a + b
    a = b
    b = c
    print(c)
```

**3) Define a class and an object**

```python
# A class in Python is a blueprint for creating objects, which are instances of the class. An object is a specific instance of a class that has its own data and methods.
# Here's a simple example to define a class and an object in Python:

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)

person = Person("Aziz", 20)
person.display_info()
```
